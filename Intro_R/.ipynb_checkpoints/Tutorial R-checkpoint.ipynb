### Missing or unknown values

- R represents missing or unknown values with the <span style="color:blue">sentinel NA</span>

- but most computations with NA will return NA

```r
NA > 0; 2.7 *NA; ! NA
```

- exception: when some identity holds for all possible inputs

```r
NA ^ 0
NA | TRUE
```

- How do we check for NA values?

```r
y <- c(4,NA,-8)
y == NA        #It does nor work, sets all to NA
```

```r
y == "NA"      #It does not work either
```

```r
is.na(y)      #This is the proper way
```

```r
y [! is.na(y)] #produce a vector with NAs removed
```

- Some built-in functions allow to skip NAs from computations
- how do we find the locations of NA values within a vector?

```r
vmv <- c(1:6,NA,NA,8:12)
#get the indexes of the values
seq(along=vmv)
# and the indexes of the NAs
seq(along=vmv)[is.na(vmv)]
#other simpler way
which(is.na(vmv))
```

- if NAs are 2 'zero count' values, we may want to replace them with 'zeros'

```r
vmv[is.na(vmv)] <- 0; vmv

vmv[which(is.na(vmv))]<- 0; vmv

ifelse(is.na(vmv),0,vmv)        #vectorized ifelse function
```

#### S3 atomic vectors: Dates-times

### Time duration

- duration represents the time difference between two pair of dates or date-times

- they are stored in <span style="color:blue">difftimes</span>

- this S3 class has a <span style="color:blue">unit</span> attribute that determines how the difference should be interpreted

```r
one_week <- as.difftime(1,unit="weeks")
attributes(one_week)
```

```r
today <- Sys.time()
next_sunday <- today + one_week

structure(next_sunday,tzone="Europe/Rome")
```

```r
fourty_min <- as.difftime(40, units="mins")
later <- today + fourty_min
later
```

### Operating on sets: union, intersect and setdiff

- given two sets, the <span style="color:blue">union()</span> function gives a set with all the elements, but counting only once those common to both sets

```r
setA <- c("a","b","c","d","e")
setB <- c("d","e","f","g")

union(setA,setB)
```

- <span style="color:blue">intersection()</span> gives only the elements they have in common

```r
intersect(setA,setB)
```

- the difference between the two sets is order-dependent

```r
setdiff(setA,setB)

setdiff(setB,setA)

#Compare if sets are equal
setequal(setA,setA)

setequal(setA,setB)

```

```r
setA %in% setB

setA[setA %in% setB]
```

### The apply() collention functions

     <span style="color:blue">apply</span>(X,MARGIN,FUN)
 
with

X: an <span style="color:blue">array or matrix</span>

MARGIN: a value <span style="color:blue">or range</span> between 1 and 2 to define where to <span style="color:blue">apply</span> the <span style="color:blue">function</span>:

MARGIN=1: the manipulation <span style="color:blue">is</span> performed <span style="color:blue">on</span> rows

MARGIN=2: the manipulation <span style="color:blue">is</span> performed <span style="color:blue">on</span> columns

MARGIN=c(1,2) the manipulation <span style="color:blue">is</span> performed <span style="color:blue">on</span> rows and columns.

FUN: <span style="color:blue">which function</span> to <span style="color:blue">apply</span>.

Built functions like <span style="color:blue">mean</span>, <span style="color:blue">median</span>, <span style="color:blue">sum</span>, <span style="color:blue">min</span>, <span style="color:blue">max</span> and user-defined functions can be applied .

### Random numbers

- random numbers from a uniform distribution $U(0,1)$ are generated using  <span style="color:blue">runif()</span>

- the random generation seed is set via  <span style="color:blue">set.seed()</span>

```r
set.seed(2021)
runif(3)
```

- resetting the seed with the same value will generate the same sequence of random numbers

- it is also possible to save the current seed and reuse it to obtain the same random numbers sub-sequence

```r
#save the current seed
current.seed <- .Random.seed
runif(3)
runif(3)

#Reset the previous sequence seed
current.seed -> .Random.seed
runif(5)
```

#### sampling from a vector

- generating random numbers from probability distributions will be discussed in the next lessons

- now we want to randomize (shuffling or sampling from) the elements of a vector

- There are two ways of sampling:

    <ol>
    <li><span style="color:blue">sampling without replacement</span>: all the vector values will appear in output, but in a randomize sequence</li>
    <li><span style="color:blue">sampling with replacement</span>: some vector values may be re-selected and appear more than once in the output</li>
    </ol>
    
- using <span style="color:blue">sample()</span>, sampling without replacement is the default operation

```r
y <- c(8,3,5,7,6,6,8,9,2,3,9,4,10,4,11)
y
#reshufling vector values
sample(y)

sample(y,5)

sample(y,5)
```

- the option <span style="color:blue">replace=T</span> allows for sampling with replacement

```r
sample(y,replace=T)
```

### Loops

- Let's create a function, using loops, to evaluate the factorial:
- $n! = n(n-1)(n-2)...2.1$

```r
fact1 <- function(x) {
    f <-1
    if (x<2) return (1)
    for (i in 2:x){
        f <- f*i
    }
    return (f)
    }

sapply(1:5,fact1)
```

```r
fact2 <- function(x) {
    f <-1; t<- x
    while (t>1) {
        f <- f*t
        t <- t-1
    }
    return (f)
    }

sapply(1:5,fact2)
```

```r
fact3 <- function(x) {
    f <-1; t<- x
    repeat {
        if (t<2) break 
        f <- f*t
        t <- t-1
    }
    return (f)
    }

sapply(1:5,fact3)
```

- But it is almost always better to use a built-in function that operates on the entire vector, removing the need of loops or repeats

```r
#it does not work for 0
cumprod(1:5)
```

```r
fact4 <- function(x) max(cumprod(1:x))
sapply(1:5,fact4)
```

```r
#R implements a factorial() function
sapply(1:5,factorial)
```

```r
y <- c(-3,4,-2,-1,8,7,9);y

for (i in 1:length(y)) {if (y[i]<0) y[i]<-0};y
```

### Loops are slow, compared to vectorized operations

- let's generate $5 \dot 10^7$ events according to an uniform distribution, $u(0,1)$

- we want to search for the maximum value in the vector using the vectorized function max() and by using conventional loops

```r
x <- runif(50000000)

system.time(max(x))

pc <- proc.time()
cmax <- x[1]
for (i in 2:length(x)) { if (x[i]>cmax) cmax <- x[i]}

proc.time()-pc
```

<span style="color:blue">system.time()</span> and <span style="color:blue">proc.time()</span> produce a vector of three numbers, showing the user, system and total elapsed time in seconds

### Good/Bad practice in building vectors

- we want to build a vector containing $ 10^n$ elements in the sequence $1:10^n$

- three ways are analyzed

```r
test1 <- function(n){
    y <- 1:n
}

test2 <- function(n){
    y <- numeric(n)
    for (i in 1:n)
        y[i] <- i
}

test3 <- function(n){
    y <- NULL
    for (i in 1:n)
        y <- c(y,i)
}
```

```r
system.time(test1(10000000))
system.time(test2(10000000))
#system.time(test3(10000000))
```

- the first method (test1) is the best

- the loop using a pre-determined vector length is reasonably fast

- the last method (test3) is the slowliest

- Moral: <span style="color:red">never grow vectors by repeated concatenation</span>
