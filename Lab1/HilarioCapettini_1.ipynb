##  Vector and Data Frames

- The following table gives the volume, area, length and maximum and mean depths of some Scottish lakes[1]. Create vectors, holding the lakeâ€™s name and all the parameters and build a dataframe called scottish.lakes from the vectors


```r
Volume     <- c(7.45,2.6,2.3,1.6,1.2,1.09,1.08,1.07,0.97,0.79,0.77,0.75,0.35)
Area       <- c(56,71,27,26.4,39,28.6,18.6,16,19,19.5,12.4,16,22.5)
Length     <- c(39,36,18.8,23,41,20,23,16,15.7,28,12.9,19.3,27.8)
Max_depth  <- c(230,190,310,150,94,114,156,162,134,128,151,109,49)
Mean_depth <- c(132,37,87,60.6,32,38,57.6,70,51,40,43.4,46.5,15.5)

lakes <- data.frame(Volume,Area,Length,Max_depth,Mean_depth,
                    row.names  = c("Loch Ness", "Loch Lomond", "Loch Morar", "Loch Tay",
                                   "Loch Awe", "Loch Maree", "Loch Ericht", "Loch Lochy",
                                   "Loch Rannoch", "Loch Shiel", "Loch Katrine", "Loch Arkaig","Loch Shin"))
                     
lakes
```

```r
lakes
```

###  evaluate the highest and lowest volume and area lake


```r
#Here we get the element index satisfying max() and print all the info of this lake
print("The lake with the highest area")
lakes[which.max(lakes$"Area"),]        

print("The lake with the smallest area")
lakes[which.min(lakes$"Area"),]
```

```r
#Here we get the element satisfying max() and print only the desired information
v_max <- which.max(lakes$"Volume")
print(paste("The lake with the highest volume is",row.names(lakes)[v_max], "with",lakes[v_max,1],"km^3"))


v_min <- which.min(lakes$"Volume")
print(paste("The lake with the smallest volume is",row.names(lakes)[v_min], "with",lakes[v_min,1],"km^3"))
```

### Order the frame with respect to the area and determine the two largest area lakes


```r
lakes <- lakes[order(lakes$"Area",decreasing=TRUE),]
print(paste("The two lakes with the largest areas are",row.names(lakes)[1],"and",row.names(lakes)[2]))

lakes[1:2,]
```

###   By summing up the areas occupied by the lakes, determine the area of Scotland covered by water


```r
area = sum(lakes$"Area")
print(paste("The total area of Scotland covered by water is",area,"km^2"))
```

### produce a plot with the Oil price as a function of the year

```r
library(showtext)
showtext_auto()

set_plot_dimensions <- function(width_choice, height_choice) {
        options(repr.plot.width=width_choice, repr.plot.height=height_choice)
        }
```

```r
set_plot_dimensions(8, 4)

plot(data_oil[,3], data_oil[,4],
    pch = 1, cex=0.5, col="navy",
    main = "Oil Price as a function of the year",
    ylab = "Price [$]", xlab ="Year")

    lines(data_oil[,3], data_oil[,4],
          col = "firebrick1", lty=1)

```

### Which is the highest price in history?  when did it occur?

```r
#Obtain the maximum price row index
p_max <- which.max(data_oil$"OilPrice")

print(paste("The highest oil price in history was", sprintf("%.1f",data_oil[p_max,4]), "$ during the year",data_oil[p_max,3]))
```

### Plot the derivative of the curve, simply evaluated with the finite difference formula (forward derivative):

 $$\frac{\partial price}{\partial year} = price_{j+1} - price_{j}$$

```r
dPdY <- vector(mode="list",length=length(data_oil[,4]))
for (i in 1:length(data_oil[,4])) {dPdY[i] <- data_oil[i+1,4]-data_oil[i,4]}
```

```r
set_plot_dimensions(8, 4)

plot(data_oil[,3],dPdY,
    pch = 1, cex=0.5, col="navy",
    main = "Derivative of Oil Price",
    ylab = "Oil Price derivative [$]", xlab ="Year")

    lines(data_oil[,3],dPdY,
          col = "grey", lty=1)
```

- for the following items select only the years â‰¥ 1970:

```r
#We keep the data for years >= 1970
data_coal_70 <- filter(data_coal, Year >= 1970)
```

```r
#We remove continents and international organisations 
banned <- c('World', 'Asia Pacific', 'Asia and Oceania', 'OECD', 'North America',
            'Eurasia','Europe','EU-28','CIS','Africa','South & Central America',
            'Central and South America','Europe (other)','Other Asia & Pacific',
            'Middle East','OPEC','Other Africa','Other CIS','Other S. & Cent. America',
            'Persian Gulf')
for (i in banned) {
    data_coal_70 <- filter(data_coal_70, Entity != toString(i))
}
```

### selecting only the year after 1970, determine the total integrated production for each country and print the top 5 countries with highest coal production

```r
countries <- distinct(data_coal_70["Entity"])
n_countries <- nrow(countries)
```

```r
ip_by_country <- matrix(ncol = 2, nrow = n_countries)

for (i in  1:n_countries) {
    filtered <- filter(data_coal_70, Entity == toString(countries[i,]))
    ip_by_country[[i,1]]     <- toString(countries[i,])
    ip_by_country[[i,2]]     <- as.double(sum(filtered[,4]))
    
}
ip_by_country <- as_tibble(ip_by_country)
```

```r
#conver the second vector to numeric vector otherwise we have problems 
# with the order function 
#https://stackoverflow.com/questions/29399518/order-function-in-r-is-not-ordering-correctly/29401459
ip_by_country$V2 <- as.numeric(as.character(ip_by_country$V2))


ip_by_country <- ip_by_country[order(ip_by_country$V2,decreasing=TRUE),]

print("The top 5 Countries with highest coal productions are")
top_5 <- ip_by_country[1:5,1]
top_5
```

### Create a plot of production vs time for those 5 countries

```r
China        <- as.data.frame(filter(data_coal_70, Entity == "China"))[,3:4]
United_tates <- as.data.frame(filter(data_coal_70, Entity == "United States"))[,3:4]
Rusia        <- as.data.frame(filter(data_coal_70, Entity == "Russia"))[,3:4]
Australia    <- as.data.frame(filter(data_coal_70, Entity == "Australia"))[,3:4]
India        <- as.data.frame(filter(data_coal_70, Entity == "India"))[,3:4]
```

```r
plot(China[,1], China[,2], pch = 7, cex=0.25, col="navy",
     xlim =c(1970,2020),
     ylim  =c(0,25000),
     xlab ="Year",
     ylab ="Coal production (TWh)",)
lines(China[,1], China[,2],col = "navy", lty=3)

lines(United_tates[,1], United_tates[,2],col = "red", lty=3)
points(United_tates[,1], United_tates[,2],col = "red", lty=3, cex=0.25)

lines(Rusia[,1], Rusia[,2],col = "pink", lty=3)
points(Rusia[,1], Rusia[,2],col = "pink", lty=3, cex=0.25)

lines(Australia[,1], Australia[,2],col = "black", lty=3)
lines(Australia[,1], Australia[,2],col = "black", lty=3, cex=0.25)

lines(India[,1], India[,2],col = "blue", lty=3)
lines(India[,1], India[,2],col = "blue", lty=3, cex=0.25)

legend(1970,25000,legend=c("China","United States","Russia","Australia","India"),
                  col=c("navy","red","pink","black","blue"),
                  lty=c(1,2,3,4,5), ncol=1)
title(main = "Production vs time for those 5 countries")
```

### generate a plot with the cumulative sum of the Worldâ€™s coal production over the years


```r
welt_cumulative <- matrix(ncol = 2, nrow = 51)

for (year in 1970:2019){
    filtered <- filter(data_coal_70, Year == year)
    i <- year-1969
    welt_cumulative[[i,1]]     <- year
    welt_cumulative[[i,2]]     <- as.double(sum(filtered[,4]))
}
```

```r
plot(welt_cumulative[,1], welt_cumulative[,2], pch = 7, cex=0.7, col="navy",
     xlim =c(1970,2020),
     #ylim  =c(0,25000),
     xlab ="Year",
     ylab ="Coal production (TWh)",)
lines(welt_cumulative[,1], welt_cumulative[,2],col = "navy", lty=3)

title(main = "World's coal production over the years")
```

```r
library(tidyverse)

library(showtext)
showtext_auto()

set_plot_dimensions <- function(width_choice, height_choice) {
        options(repr.plot.width=width_choice, repr.plot.height=height_choice)
        }
```

```r
id <- 'https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/vaccinations/vaccinations-by-manufacturer.csv'
data_covid <- read_csv(id)
```

```r
data_covid[100:105,]
```

###  filter() the original tibble by selecting the following countries: Italy


```r
data_covid_italy <- filter(data_covid, location == "Italy")
data_covid_italy[1:3,]
```

### plot the number of vaccines given as a function of time for the different vaccine manufacturer


```r
#We add a column for the number of vaccines given every day by each company  
data_covid_italy <- data_covid_italy %>%
    group_by(vaccine)%>%
    arrange(vaccine) %>%
    mutate(diff = total_vaccinations - lag(total_vaccinations, default = first(total_vaccinations)))
```

```r
data_covid_italy[1:3,]
```

```r
set_plot_dimensions(8, 4)
ggplot(data_covid_italy, aes(date, diff, colour = vaccine, shape = vaccine )) +
       geom_line(size = 0.2)+ geom_point(size=0.7)+
       labs(title ="Number of vaccines/day given as a function of time for the different vaccine manufacturer")+
       ylab("Vaccines") + xlab("Date")+scale_y_continuous(labels = scales::scientific)
```

### from the same tibble plot the total number of vaccines shot per day in Italy


```r
#https://uomresearchit.github.io/r-tidyverse-intro/04-dplyr/
```

```r
total_i <- data_covid_italy %>%
  group_by(date) %>% 
  summarise(vaccines = sum(diff))
```

```r
set_plot_dimensions(8, 4)

ggplot(total_i, aes(date, vaccines)) +
       geom_line(size = 0.2)+ geom_point(size=0.7) +
       labs(title ="Total number of vaccines shot per day in Italy.")+
      ylab("Vaccines") + xlab("Date") +scale_y_continuous(labels = scales::scientific)
```
